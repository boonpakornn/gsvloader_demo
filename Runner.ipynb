```python
geojson_prop = {
    'PNAME': 'กรุงเทพมหานคร'
}
entry = {
    'GEOJSON_PATH': 'geojson/subdist.geojson',
    'width': 640,
    'height': 500,
    'degrees': [k*72 for k in range(int(360/72))], # [0, 72, 144, 216, 288]
    'fov': 75,
    'pitch': -15,
    'keys': keys
}
```

```python
for ind in data_entry:
    data = data_entry[ind]
    if(int(ind) < 148):
        continue

    entry['OUTPUT_IMAGE_PATH'] = os.path.join('กรุงเทพมหานคร', data['DNAME'], data['SNAME'], 'original')
    if(os.path.exists(entry['OUTPUT_IMAGE_PATH'])):
        print('Index: ' + ind +' '+data['DNAME'], data['SNAME'])
        continue
    print('===================\n\nIndex: ' + ind +' '+data['DNAME'], data['SNAME'])
    geojson_prop['DNAME'] = data['DNAME']
    geojson_prop['SNAME'] = data['SNAME']
    points = polygon_to_points.run(entry, geojson_prop)
    loader.run(entry, points)
```
